```python
import pandas as pd
import random
import numpy as np
```

```python
# Setting seed for reproducibility
random.seed(42)
```

```python
# Generate data
num_rows = 1000
```

```python
# Customer ID - unique
customer_ids = random.sample(range(1000000000, 9999999999), num_rows)

# Income level - range from 20,000 to 500,000
incomes = [random.randint(20000, 500000) for _ in range(num_rows)]

# Age group - range from 16 to 90
ages = [random.randint(16, 90) for _ in range(num_rows)]

# Household size - range from 1 to 6
household_sizes = [random.randint(1, 6) for _ in range(num_rows)]

# Shopping frequency per month - range from 1 to 10
shopping_frequencies = [random.randint(1, 10) for _ in range(num_rows)]

# Region/store location - North/South/East/West
regions = random.choices(['North', 'South', 'East', 'West'], k=num_rows)

# Purchase preferences - one of the given categories
purchase_preferences = random.choices(
    ['Fresh Produce', 'Bakery', 'Deli', 'Meat and Seafood', 'Dairy', 'Frozen Foods', 'Packaged Foods', 'Cereal and Breakfast Foods'], 
    k=num_rows
)

# Create DataFrame
df_random_data = pd.DataFrame({
    "Customer ID": customer_ids,
    "Income": incomes,
    "Age": ages,
    "Household Size": household_sizes,
    "Shopping Frequency per month": shopping_frequencies,
    "Region/Store Location": regions,
    "Purchase Preferences": purchase_preferences
})

# Display the DataFrame
print(df_random_data.head())  
```
